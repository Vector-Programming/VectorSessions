```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import random

%matplotlib inline
# plt.rcParams["figure.figsize"] = [17, 7]
```

```python
d_size = 100
d_noize = 1

x = np.random.randn(d_size, 1)
y = 2*x + 3 + d_noize*np.random.randn(d_size, 1)
```

```python
pred = LinearRegression()
pred.fit(x, y)
```

```python
f, axarr = plt.subplots(1, 2, sharey=True)

axarr[0].scatter(x, y, s=5)
axarr[0].legend(['Data'])
axarr[0].set_title('Data Points')

axarr[1].scatter(x, y, s=5)
axarr[1].plot(x, pred.predict(x), linewidth=0.3, color='green')
axarr[1].legend(['Linear Model', 'Data'])
axarr[1].set_title('Linear Model')

plt.show()
```
