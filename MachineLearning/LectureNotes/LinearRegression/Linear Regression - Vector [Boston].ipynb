```python
import matplotlib.pyplot as plt # visualization
import numpy as np # matrix ops.
import pandas as pd # data analysis
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
boston = load_boston() # load the boston data set into a variable
```

```python
# seperate out the features and the labels
features = np.array(boston.data)
labels = np.array(boston.target)
```

```python
pd.DataFrame(features).describe() # do some data analysis on the features
```

```python
pd.DataFrame(labels).describe() # do some data analysis on the labels
```

## Theory
In statistics, linear regression is a linear approach for modelling the relationship between a scalar dependent variable y and one or more explanatory variables (or independent variables) denoted X. The case of one explanatory variable is called simple linear regression. For more than one explanatory variable, the process is called multiple linear regression.<img src="https://3qeqpr26caki16dnhd19sv6by6v-wpengine.netdna-ssl.com/wp-content/uploads/2017/12/Scatter-Plot-of-SVD-Solution-to-the-Linear-Regression-Problem.png" style="width: 500px;">

### In Simple Terms
Take a line with the equation `y = ax + b` in the coordinate axis, where:
- a is the slope of the line
- b is the X-Intercept of the line

We are basically looking at the points, and finding the values of `a` and `b` to make the best possible **line of fit**.

The power of Python and its community is displayed here as we do not need to understand and implement the mathematics and the calculations needed to find these optimal values. **sklearn** has just what we need in their modules.

```python
reg = LinearRegression() # create the linear regression model
```

```python
# prepare and split the data into the specific variables and convert to np arrays for convinience
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
from time import time
t = time()
reg.fit(x_train, y_train) # "fit" the Linear Model
print(round(time()-t, 3), 's')
```

```python
predictions = np.array(reg.predict(x_test)) # get the predictions of the model for the test data
```

```python
predictions[:5] # print out the predictions
```

```python
y_test[:5] # print out the correct values, compare with the predictions
```

```python
print("Accuracy Of Model: {}%".format(np.mean((predictions-y_test)**2))) # calculate the accuracy of the model
```

```python
#plot the various dimensions and linear equations
plt.rcParams["figure.figsize"] = [17, 7]
n = 506
ndims = 13

for i in range(ndims):
    plt.scatter(x_test[:n, i], y_test[:n])
    plt.plot(x_test[:n, i], predictions[:n])
plt.show()
```
